```python
def rename_opposition(x):
    if x == 'sri':
        return 'sri lanka'
    elif x == 'new':
        return 'new zealand'
    elif x == 'west':
        return 'west indies'
    elif x == 'south':
        return 'south africa'
    elif x == 'hong':
        return 'hong kong'
    else:
        return x
```

```python
data['opposition'] = data['opposition'].apply(rename_opposition)
```

```python
def convert_to_int(x):
    if not pd.isna(x):
        return int(x)
    else:
        return x
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

data[['opposition']] = encoder.fit_transform(data[['opposition']])
```

```python
# Create a dictionary to store individual DataFrames
player_dataframes = {}

# Iterate through unique player IDs
for player_id, player_name in data[['player_id', 'player_name']].drop_duplicates().values:
    # Filter the DataFrame for the current player ID
    player_df = data[data['player_id'] == player_id]
    player_name = player_name.replace(" ", "_")
    # Store the player-specific DataFrame in the dictionary with the player's name as the key
    player_dataframes[player_name] = player_df

# Access individual player DataFrames by player name
for player_name, player_df in player_dataframes.items():
    print(f"Player Name: {player_name}")
    print(player_df)
    print("\n")
```
